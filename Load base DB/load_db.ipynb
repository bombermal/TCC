```python
# CustomerMgmt
# Load xml file
file_tree = ET.parse(f"{actual_path}/CustomerMgmt.xml")
root = file_tree.getroot()
```

```python
class Action:
    def __init__(self, action_type, action_ts):
        self.action_type = action_type
        self.action_ts = action_ts
        
    def print_all(self):
        return [self.action_type, self.action_ts, self.c_id]
       
class Inact(Action):
    def __init__(self, action_type, action_ts, c_id):
        super().__init__(action_type, action_ts)
        self.c_id = c_id

for customer_child in root[4000:4010]:
    # Customer
    act_tp = customer_child.get("ActionType")        
    act_ts = customer_child.get("ActionTS")
    
    if act_tp == "NEW":
        pass
    elif act_tp == "ADDACCT":
        pass
    elif act_tp == "UPDCUST":
        pass
    elif act_tp == "UPDACCT":
        pass
    elif act_tp == "CLOSEACCT":
        pass
    elif act_tp == "INACT":
        obj = Inact(act_tp, act_ts, customer_child[0].attrib["C_ID"])
        print(obj.print_all())
    else:
        print("Action Type not found")
```

```python
# Finwire - Find all files names and sort them
fin_files = os.listdir(f"{actual_path}")
fin_files = filter(lambda y: "audit" not in y, filter(lambda x: "FINWIRE" in x, fin_files))
fin_files = sorted(fin_files, reverse=False)
```

```python
# Store all files in a single temporary dataframe to be splited later
# The finwire files are separated in three types CMP, FIN and SEC
df_temp = pd.DataFrame(columns=["temp_col"])
for file in fin_files:
    df_temp = pd.concat( [df_temp, pd.read_csv(f"{actual_path}{file}", header=None, names=["temp_col"])], axis=0)
df_temp.reset_index(drop=True, inplace=True)
df_temp["type"] = df_temp["temp_col"].apply(lambda x: x[15:18])
df_temp.head()
```

```python
cmp_cols = ["PTS", "RecType", "CompanyName", "CIK", "Status", "IndustryID", "SPrating", "FoundingDate", "AddrLine1", "AddrLine2"
            , "PostalCode", "City", "StateProvince", "Country", "CEOname", "Description"]
sec_cols = ["PTS", "RecType", "Symbol", "IssueType", "Status", "Name", "ExID", "Shout", "FirstTradeDate", "FirstTradeExchg",
            "Dividend", "CoNameOrCIK"]
fin_cols = ["PTS", "RecType", "Year", "Quarter", "QtrStartDate", "PostingDate", "Revenue", "Earnings", "EPS", "DilutedEPS",
            "Margin", "Inventory", "Assets", "Liabilities", "Shout", "DilutedShOut", "CoNameOrCIK"]

# Use boolean masks to split the dataframe
cmp_mask = df_temp.type == "CMP"
sec_mask = df_temp.type == "SEC"
fin_mask = df_temp.type == "FIN"

# Store each dataframe in a different variable
# Convert dataframe to fwf
cols_widths = [15, 3, 60, 10, 4, 2, 4, 8, 80, 80, 12, 25, 20, 24, 46, 150]
accum = 0
accum_index = []
for idx in cols_widths:
    v2 = accum+idx
    accum_index.append((accum, v2))
    accum = v2
    
df_finwire_cmp = pd.DataFrame(
    df_temp.loc[cmp_mask, "temp_col"].copy().map(
        lambda row: [ row[x:y].strip() for x, y in accum_index ])
    .tolist(), columns=cmp_cols)

cols_widths = [15, 3, 15, 6, 4, 70, 6, 13, 8, 8, 12, 60]
accum = 0
accum_index = []
for idx in cols_widths:
    v2 = accum+idx
    accum_index.append((accum, v2))
    accum = v2
    
df_finwire_sec = pd.DataFrame(
    df_temp.loc[sec_mask, "temp_col"].copy().map(
        lambda row: [ row[x:y].strip() for x, y in accum_index ])
    .tolist(), columns=sec_cols)

cols_widths = [15, 3, 4, 1, 8, 8, 17, 17, 12, 12, 12, 17, 17, 17, 13, 13, 60]
accum = 0
accum_index = []
for idx in cols_widths:
    v2 = accum+idx
    accum_index.append((accum, v2))
    accum = v2
    
df_finwire_fin = pd.DataFrame(
    df_temp.loc[fin_mask, "temp_col"].copy().map(
        lambda row: [ row[x:y].strip() for x, y in accum_index ])
    .tolist(), columns=fin_cols)
```
